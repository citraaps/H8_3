# Naive Bayes Classifier Building in Scikit-learn

```python
weather = ['Sunny', 'Sunny', 'Overcast', 'Rainy', 'Rainy', 'Rainy',
          'Overcast', 'Sunny', 'Sunny', 'Rainy', 'Sunny', 'Overcast',
          'Overcast', 'Rainy']

temp = ['Hot', 'Hot', 'Hot', 'Mild', 'Cool', 'Cool', 'Cool', 'Mild',
       'Cool', 'Mild', 'Mild', 'Mild', 'Hot', 'Mild']

play = ['No', 'No', 'Yes', 'Yes', 'Yes', 'No', 'Yes', 'No', 'Yes',
       'Yes', 'Yes', 'Yes', 'Yes', 'No']
```

```python
from sklearn import preprocessing

le = preprocessing.LabelEncoder()

weather_encoded = le.fit_transform(weather)

print(weather_encoded)
```

```python
temp_encoded = le.fit_transform(temp)

label = le.fit_transform(play)

print(temp_encoded)
print(label)
```

```python
features = list(zip(weather_encoded, temp_encoded))

features
```

```python
from sklearn.naive_bayes import GaussianNB

model = GaussianNB()

model.fit(features, label)

predicted = model.predict([[0,2]])
print("Predicted Value:", predicted)
```

# Naive Bayes with Multiple Labels

```python
from sklearn import datasets

wine = datasets.load_wine()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(wine.data, wine.target,
                                                   test_size=0.3, random_state=109)
```

```python
gnb = GaussianNB()

gnb.fit(X_train, y_train)

y_pred = gnb.predict(X_test)
```

```python
from sklearn import metrics

metrics.accuracy_score(y_test, y_pred)
```

# Decision Tree Classifier

```python
import pandas as pd

from sklearn.tree import DecisionTreeClassifier
```

```python
col_names = ['pregnant', 'glucose', 'bp', 'skin', 'insulin',
            'bmi', 'pedigree', 'age', 'label']

pima = pd.read_csv('https://raw.githubusercontent.com/ardhiraka/PFDS_sources/master/diabetes.csv', header=None, names=col_names)
```

```python
pima.head()
```

```python
pima.info()
```

```python
numer = ['pregnant', 'insulin', 'bmi', 'age', 'glucose', 'bp', 'pedigree', 'label']

for col in numer:
    pima[col] = pd.to_numeric(pima[col], errors='coerce')
```

```python
pima.dropna(inplace=True)
```

```python
feature_cols = ['pregnant', 'insulin', 'bmi', 'age', 'glucose', 'bp', 'pedigree']

X = pima[feature_cols]
y = pima.label
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
clf = DecisionTreeClassifier()

clf = clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)
```

```python
metrics.accuracy_score(y_test, y_pred)
```

# Random Forest

```python
iris = datasets.load_iris()
```

```python
iris.target_names
```

```python
iris.feature_names
```

```python
iris.data[:5]
```

```python
data = pd.DataFrame({
    'sepal length':iris.data[:,0],
    'sepal width':iris.data[:,1],
    'petal length':iris.data[:,2],
    'petal width':iris.data[:,3],
    'species':iris.target
})

data.head()
```

```python
X = data[['sepal length', 'sepal width', 'petal length', 'petal width']]
y = data['species']


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(n_estimators=100)

clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)
```

```python
metrics.accuracy_score(y_test, y_pred)
```

```python
clf.predict([[3, 5, 4, 2]])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

sns.barplot(x=feature_imp, y=feature_imp.index)

plt.show()
```

# Exercise: Multiple Algorithm on Dataset

```python
import numpy as np
```

```python
train = pd.read_csv('https://raw.githubusercontent.com/ardhiraka/PFDS_sources/master/Final_Dataset/train.csv')
```

```python
train.head()
```

```python
train.describe()
```

```python
train.isnull().sum()
```

```python
train.fillna(train.mean(), inplace=True)
train.isnull().sum()
```

```python
train.Gender.fillna(train.Gender.mode()[0], inplace=True)
train.Married.fillna(train.Married.mode()[0], inplace=True)
train.Dependents.fillna(train.Dependents.mode()[0], inplace=True)
train.Self_Employed.fillna(train.Self_Employed.mode()[0], inplace=True)
train.isnull().sum()
```

```python
train.Loan_Amount_Term = np.log(train.Loan_Amount_Term)
```

```python
X = train.drop('Loan_Status', 1)
y = train.Loan_Status
```

```python
X = pd.get_dummies(X)
train = pd.get_dummies(train)
```

```python
X.head()
```

```python
y.head()
```

```python
x_train, x_cv, y_train, y_cv = train_test_split(X, y, test_size=0.2)
```

- Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()

model.fit(x_train, y_train)
```

```python
pred_cv = model.predict(x_cv)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

accuracy_score(y_cv, pred_cv)
```

```python
confusion_matrix(y_cv, pred_cv)
```

- Decision Tree Algorithm

```python
dt = tree.DecisionTreeClassifier()
dt.fit(x_train, y_train)
```

```python
pred_cv1 = dt.predict(x_cv)
```

```python
accuracy_score(y_cv, pred_cv1)
```

```python
confusion_matrix(y_cv, pred_cv1)
```

- Random Forest

```python
rf = RandomForestClassifier()
rf.fit(x_train,y_train)
```

```python
pred_cv2 = rf.predict(x_cv)
```

```python
print(accuracy_score(y_cv, pred_cv2))
print(confusion_matrix(y_cv, pred_cv2))
```

- SVM

```python
svm_model = svm.SVC()
svm_model.fit(x_train, y_train)
```

```python
pred_cv3 = svm_model.predict(x_cv)
```

```python
print(accuracy_score(y_cv, pred_cv3))
print(confusion_matrix(y_cv, pred_cv3))
```

- Naive Bayes

```python
nb = GaussianNB()
nb.fit(x_train,y_train)
```

```python
pred_cv4 = nb.predict(x_cv)
```

```python
print(accuracy_score(y_cv, pred_cv4))
print(confusion_matrix(y_cv, pred_cv4))
```
